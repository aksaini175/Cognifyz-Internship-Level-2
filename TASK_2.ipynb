dentify The Most Common Combinations of Cuisines in The Dataset

```python

df.columns = df.columns.str.strip()


df = df.dropna(subset=['Cuisines'])

common_cuisine_combinations = df['Cuisines'].value_counts().head(10)

print("Top 10 Most Common Cuisine Combinations:\n")
print(common_cuisine_combinations)
```

```python
common_cuisine_combinations.reset_index().rename(
    columns={'index': 'Cuisine Combination', 'Cuisines': 'Count'}
)
```

```python
print("Top 10 Most Common Cuisine Combinations:")
print(common_cuisine_combinations)
```

Determine if Certain Cuisine Combinations tend to Have Higher **Ratings**

```python
import pandas as pd


df.columns = df.columns.str.strip()

df = df.dropna(subset=['Cuisines', 'Aggregate rating'])
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')

cuisine_ratings = (
    df.groupby('Cuisines')['Aggregate rating']
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

print("Top 10 Cuisines by Average Rating:\n")
print(cuisine_ratings.round(2))
```

```python
print("\nTop 10 Highest-Rated Cuisine Combinations:")
print(cuisine_ratings)
```
