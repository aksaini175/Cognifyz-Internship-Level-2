```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/Dataset .csv')
```

Analyze The Distribution of Aggregate Ratings and Determine The Most Common Rating **Range**

```python
import pandas as pd

df.columns = df.columns.str.strip()

df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')


aggregate_ratings = df['Aggregate rating']
rating_counts = aggregate_ratings.value_counts().sort_index()

rating_counts
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df.columns = df.columns.str.strip()


aggregate_ratings = pd.to_numeric(
    df['Aggregate rating'],
    errors='coerce'
).dropna()

# Create histogram
plt.figure(figsize=(8, 5))
plt.hist(aggregate_ratings, bins=20, edgecolor='black')

plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df.columns = df.columns.str.strip()

df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')

rating_counts = df['Aggregate rating'].dropna().value_counts().sort_index()

most_common_rating = rating_counts.idxmax()

print("The most common aggregate rating is:", most_common_rating)
```

The Average Number of Votes Received by **Restaurants**

```python
import pandas as pd


df.columns = df.columns.str.strip()


df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')

average_votes = df['Votes'].mean()

print(f"The average number of votes received by restaurants is {average_votes:.2f}")
```
